```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
# ----Cargo la imagen -----

img = cv2.imread('img_calculadora.tif',cv2.IMREAD_GRAYSCALE) 
plt.figure(), plt.imshow(img, cmap='gray'), plt.show(block=False)
```

```python
# Determinar las dimensiones de la imagen 
w,h = img.shape

print(f"Altura: {h}")
print(f"Ancho: {w}")
```

```python
# Determinar el valor máximo y mínimo de nivel de gris de img_calculadora.tif

print(img.min(), img.max())
```

```python
# Buscar todos los valores de nivel de gris que tiene la imagen, cuantos son? 

pix_vals = np.unique(img)
N_pix_vals = len(np.unique(img))

print(pix_vals)
print(N_pix_vals)
```

```python
# Recortar de la imagen las teclas SIN, COS y TAN y mostrarlas en tres subplots distintos

img_copy = img.copy()

sen = img[340:430,750:860] 
cos = img[340:430,975:1085] 
tan = img[340:430,1200:1310]

# Crear subplots
fig, axes = plt.subplots(1, 3, figsize=(10, 5))

# Mostrar la imagen original en el primer subplot
axes[0].imshow(sen, cmap='gray')
axes[0].axis('off')  # Ocultar los ejes

# Mostrar la imagen modificada 1 en el segundo subplot
axes[1].imshow(cos, cmap='gray')
axes[1].axis('off')  # Ocultar los ejes

# Mostrar la imagen modificada 2 en el tercer subplot
axes[2].imshow(tan, cmap='gray')
axes[2].axis('off')  # Ocultar los ejes

# Mostrar los subplots
plt.show()
```

```python
# Pegar los recortes anteriores en la imagen original en el orden TAN COS SIN

img_copy[340:340+sen.shape[0], 1200:1200+sen.shape[1]] = sen
img_copy[340:340+tan.shape[0], 750:750+tan.shape[1]] = tan


plt.figure(), plt.imshow(img_copy, cmap='gray'), plt.show(block=False)
```
